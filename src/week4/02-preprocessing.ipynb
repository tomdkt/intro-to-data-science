### Let's see how to perform a shift
#### Let's import Apple's stock data
#### https://finance.yahoo.com/quote/AAPL/history/

```python
import yfinance as yf
from datetime import datetime

# Download Apple stock data
df2 = yf.download('AAPL', start='2020-01-01', end='2024-12-31', auto_adjust=False)
df2.head()
```

```python
df2[('Close_yesterday', 'AAPL')] = df2[('Close', 'AAPL')].shift(1) # MultiIndex
print(df2[['Close', 'Close_yesterday']].head(10))
```

### New column with the day-over-day percentage change

```python
df2['change %'] = ((df2[('Close', 'AAPL')] / df2[('Close_yesterday', 'AAPL')] - 1)) * 100
df2.head()
```

```python
df2.describe(include= 'all')
```

### See the scikit-learn documentation on preprocessing!!!!

```python
from sklearn.preprocessing import LabelEncoder
```

```python
countries=df['winner_ioc']
countries
```

```python
labelEncoder = LabelEncoder()
countries_numbers = labelEncoder.fit_transform(countries)
countries_numbers
```

```python
df['countries_numbers']=countries_numbers
df.head()
```

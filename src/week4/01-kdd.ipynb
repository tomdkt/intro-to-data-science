### Generate a Pearson correlation matrix for the variables
#### Note: A value of 1 means a perfect positive correlation, -1 is a perfect negative correlation,
#### and values close to zero indicate little to no correlation.

```python
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=[np.number])

# Create upper triangle mask for correlation heatmap
mask = np.triu(np.ones_like(numeric_df.corr(), dtype=bool))
```

```python
# Set matplotlib to use a standard font
plt.rcParams['font.family'] = 'DejaVu Sans'
# OR use seaborn's default style which handles fonts well
sns.set_style("whitegrid")

plt.figure(figsize=(20,20))
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), mask=mask, square=True, annot=True, vmin=-1, vmax=1)

plt.show()
```

```python
# Open file (let's look at its contents, it's text information copied from Wikipedia - https://en.wikipedia.org/wiki/Machado_de_Assis)"
file = open('machado_de_assis.txt','r')
content = file.readlines()
file.close()

print(content)
print('-----')
for line in content:
    print(line)
```

```python
x=''
for line in content:
    if line!='':
        x+=line

print(x)
x=x.replace('Rio de Janeiro','Ribeirão Preto')
print('--------')
print(x)
```
